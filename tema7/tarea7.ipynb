```r
# https://stackoverflow.com/q/8047668/3113008
    makeTransparent<-function(someColor, alpha=100){
    newColor<-col2rgb(someColor)
    apply(newColor, 2, function(curcoldata){rgb(red=curcoldata[1], green=curcoldata[2],
    blue=curcoldata[3],alpha=alpha, maxColorValue=255)})
}

f = function(x, o){
    r <- switch(o,   
        "1" = return(1 / x),
        "2" = return(x ** 2),
        "3" = return(x ** 3),
        "4" = return(log(x)),
        "5" = return(exp(x)),
        "6" = return(sin(x * 180 / pi)),
        "7" = return(cos(x * 180 / pi)),
        "8" = return(tan(x * 180 / pi)),
    )
}

tuckey = function(x, l){
    if (l > 0){
        return(x ** l)
    } else if (l < 0){
        return(-(x ** l))
    } else {
        return(log(x))
    }
}

boxcox = function(x, l){
    return (((x ** l) - 1) / l)
    #return ( (exp(l * log(x)) - 1) / l )
}
```

```r
gifs = T

x = runif(1000, -100, 100) #seq(-100, 100, 0.5)
ls = seq(-3, 3, 0.5)
ls_str = rep(ls, 2)
ls_str = c(ls_str[1:6], ls_str[8:19], ls_str[21:26])

os = c("1 / x", "x ** 2", "x ** 3", "log(x)", "exp(x)", "sin(x * 180 / pi)", "sin(x * 180 / pi)", "sin(x * 180 / pi)") #operaciones

ts = c("xT", "yT", "xT_yT")
for (t in ts){
    for (j in 1:8){
        if (t != "xT" && j == 4){
            next
        }
        y = f(x, j)
        df = data.frame(x, y)
        df = df[complete.cases(df), ]
        df = df[!is.infinite(rowSums(df)),]
        pdf(paste(t, "func", j, ".pdf", sep = ""))
        plot(df, pch= 16)
        dev.off()

        corr_res = c()

        for (i in 1:length(ls)){
            if (ls[i] == 0){
                next
            }
            
            if (t == "xT"){
                x1 = tuckey(df['x'], ls[i])
                y1 = df['y']
            } else if (t == "yT"){
                x1 = df['x']
                y1 = tuckey(df['y'], ls[i])
            } else {
                x1 = tuckey(df['x'], ls[i])
                y1 = tuckey(df['y'], ls[i])
            }
            
            df1 = data.frame(x1, y1)
            df1 = df1[complete.cases(df1), ]
            # https://stackoverflow.com/a/36590347/3113008
            df1 = df1[!is.infinite(rowSums(df1)),]
            c1 = cor(x1, y1)
            corr_res = c(corr_res, c1)
            
            if (gifs){
                png(sprintf("imgs/%02d.png", i))
                plot(df1, pch= 16)
                legend("top", legend=c(t, os[j], "Tuckey", ls[i], paste("lambda = ", ls[i]), paste("corr = ", round(c1, 2))))
                dev.off()
            }
        }

        for (i in 1:length(ls)){
            if (ls[i] == 0){
                next
            }
            
            if (t == "xT"){
                x2 = boxcox(df['x'], ls[i])
                y2 = df['y']
            } else if (t == "yT"){
                x2 = df['x']
                y2 = boxcox(df['y'], ls[i])
            } else {
                x2 = boxcox(df['x'], ls[i])
                y2 = boxcox(df['y'], ls[i])
            }
            
            df2 = data.frame(x2, y2)
            df2 = df2[complete.cases(df2), ]
            # https://stackoverflow.com/a/36590347/3113008
            df2 = df2[!is.infinite(rowSums(df2)),]
            c2 = cor(x2, y2)
            corr_res = c(corr_res, c2)

            if (gifs){
                png(sprintf("imgs/%02d.png", (13 + i)))
                plot(df2, pch= 16)
                legend("top", legend=c(t, os[j], "Box-Scott", ls[i], paste("lambda = ", ls[i]), paste("corr = ", round(c2, 2))))
                dev.off()
            }
        }
        
        if (gifs){
            s = paste("magick -delay 100 imgs/*.png ", t, j, ".gif", sep="")
            shell(s)
            shell('echo s | del /S "imgs/"')
        }

        pdf(paste(t, "corr", j, ".pdf", sep = ""))
        plot(corr_res, xaxt="n", ylim=c(-1, 1), xlab = "", ylab ="CorrelaciÃ³n")
        rect(0, par("usr")[3], 12.5, par("usr")[4], col = makeTransparent("red"))
        rect(12.5, par("usr")[3], 26, par("usr")[4], col = makeTransparent("blue"))
        axis(1, at=1:length(ls_str), labels=ls_str, las=2)
        legend("bottomright", legend=c('Tuckey', 'Box-Cox'), cex=1.2, fill= c("red", "blue"))
        dev.off()
    }
}
```
