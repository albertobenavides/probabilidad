```r
x = 1:500 / 10
y = cos(x)
x = sin(x)
```

```r
plot(convolve(x, y, type = 'open'))
par(new=TRUE)
plot(x, type = 'l')
par(new=TRUE)
plot(y, type = 'l')
```

```r
plot(scale(convolve(y, x, type = 'circular')))
lines(x)
lines(y)
```

```r
X = runif(1000)
Y = rbinom(1000, 100, 0.05)
```

```r
pdf("X.pdf")
hist(X, main=NULL, ylab="Frecuencia")
dev.off()

pdf('Y.pdf')
hist(Y, main=NULL, ylab="Frecuencia")
dev.off()
```

```r
a = sample(-10:10, 1)
b = sample(-10:10, 1)
c = sample(-10:10, 1)
d = sample(-10:10, 1)

cov(a * X + b, c * Y + d) == a * c * cov(X, Y)
```

```r
var(X + Y) == var(X) + var(Y) + 2 * cov(X, Y)
```
